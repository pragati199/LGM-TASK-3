```python
df.rename(columns={'iyear':'Year','imonth':'Month','city':'City','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
df['Casualities'] = df.Killed + df.Wounded
df=df[['Year','Month','Day','Country','Region','City','latitude','longitude','AttackType','Killed','Wounded','Casualities','Target','Group','Target_type','Weapon_type']]
df.head(10)
```

### Data Visualization

```python
### yearly count of terrorist attack           ### year


plt.figure(figsize=(15,10))
sn.countplot(x="Year", data=df,color = 'green')
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show() 
```

### Observation
From the graph we can see 2013-17 marks the highest attacks with 2014 having the highest.

There has been a gradual increase in Terror Activities since 2004.

Highest number of terror activities occurred in the year 2014.

After 2014 the terror activities started to decrease.

```python
### month analysis            ### month


df['Month'].value_counts()
plt.figure(figsize = (9,5))
sn.countplot(x='Month', data = df,color = 'cyan')
plt.xticks(rotation=90)
plt.show()
```

```python
### Terrorist Activities By Region In Each Year


pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(18,8))
plt.title('Terrorist Activities By Region In Each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
df['casualities'] = df['Killed'] + df['Wounded']
```

```python
terror1 = df.sort_values(by='casualities',ascending=False)[:40]
```

```python
df1=terror1.pivot_table(index='Country',columns='Year',values='casualities')
df1.fillna(0,inplace=True)
```

```python
df1.head(10)
```

```python
### Region affected by terrorist attack               ### region

df['Region'].value_counts()
plt.figure(figsize=(12,6))
sn.countplot(x='Region',data=df)
plt.xticks(rotation=70)
plt.show()
```

```python
### countries affectedby terrorist attacks             ### countries

df.Country.value_counts()
```

```python
df.Country.unique()
```

```python
df.Country.value_counts()[:10].plot.bar(figsize=(12,6))            ### country
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=70)
plt.title('Top 10 Countries Affected by Terrorism')
plt.show()
```

```python
df.City.value_counts()[:15].plot.bar(figsize = (15,7),color = 'brown')          ### cities
plt.xlabel('Cities')
plt.ylabel('Count')
plt.title('Top cities affected by terrorism')
plt.show()
```

```python
df.groupby('City')['Country'].size()[:15].plot.pie(figsize = (25,10))
plt.title('Top 15 cities affected by terrorism')
plt.ylabel(' ')
plt.show()
```
